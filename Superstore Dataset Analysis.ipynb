```python
# total profit by state

tps = df.groupby('State')['Profit'].sum().sort_values(ascending=False).reset_index()
tps.head().style.background_gradient(cmap='coolwarm_r')
```

```python
tps.plot(x='State', y='Profit', kind='bar', figsize=(15, 7), color='green')
```

```python
# average profit by state

aps = df.groupby('State')['Profit'].mean().sort_values(ascending=False).reset_index()
aps.head()

aps.plot(x='State', y='Profit', kind='bar', figsize=(15, 7), color='orange')
```

```python
# avg discount by state

ads = df.groupby('State')['Discount'].mean().sort_values(ascending=False).reset_index()
ads.head()

ads.plot(x='State', y='Discount', kind='bar', figsize=(15, 7), color='red')
```

```python
# order date vs shipment date

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# order date and shipment date comparison

os = df.groupby('Order Date')['Ship Date'].count().reset_index()
os.head()
```

```python
df.columns
```

```python
# ship mode vs segment

ss = df.groupby(['Ship Mode', 'Segment'])['Sales'].sum().reset_index()
ss

ss.plot(x='Ship Mode', y='Sales', kind='bar', figsize=(15, 7), color='green')
plt.ylabel('Sales')
plt.show()
```

```python
# Region and avg discount

rd = df.groupby('Region')['Discount'].mean().reset_index()
rd

rd.plot(x='Region', y='Discount', kind='bar', figsize=(15, 7), color='orange')
plt.ylabel('Discount')
plt.show()
```

```python
# total loss per state
ls = df.groupby('State')['Profit'].sum().sort_values(ascending=False).reset_index()
ls

ls.plot(x='State', y='Profit', kind='bar', figsize=(15, 7), color='red')
plt.ylabel('Loss')
plt.show()

```

```python
tps = df.groupby('State')['Profit'].sum().sort_values(ascending=False).reset_index()

loss = tps[tps['Profit'] < 0]

print(loss)

loss.plot(x='State', y='Profit', kind='bar', figsize=(15, 7), color='red')
plt.title('Loss per State') # Added title
plt.ylabel('Profit') # Corrected ylabel
plt.show()
```
