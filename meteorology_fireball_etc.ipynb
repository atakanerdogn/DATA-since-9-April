```python
# remove outliers

q1 = num.quantile(0.25)
q3 = num.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
num = num[~((num < (q1 - 1.5 * iqr)) | (num > (q3 + 1.5 * iqr))).any(axis=1)]

num.describe()
```
