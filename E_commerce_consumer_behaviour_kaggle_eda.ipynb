```python
e_num = e.select_dtypes(include=np.number)
e_num.head()
```

```python
e_cat = e.select_dtypes(include='object')
e_cat.head()
```

```python
# to check if we should drop the customer ids
customer_id = e['Customer_ID'].value_counts().reset_index()
customer_id
```

```python
e.drop(columns='Customer_ID',axis=1,inplace=True)
e.head()
```
