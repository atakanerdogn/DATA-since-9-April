```python
q1 = df['Employer_Reputation_Score'].quantile(0.25)
q2 = df['Employer_Reputation_Score'].quantile(0.5)
q3 = df['Employer_Reputation_Score'].quantile(0.75)
iqr = q3 - q1

removed_outliers = df[(df['Employer_Reputation_Score'] < q1 - 1.5 * iqr) | (df['Employer_Reputation_Score'] > q3 + 1.5 * iqr)]
removed_outliers

df_new = removed_outliers.to_csv('new_data.csv', index=False)
```
